### Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
ratings = pd.read_csv('F:/Обучение/Профессия Аналитик данных/Python/DataSet/ml-latest/ratings.csv')
ratings.head()
```

```python
ratings_movies = ratings.groupby('movieId').agg({'rating':'mean'})
ratings_movies.head()
```

```python
def class_movies(row):
    if row.rating > 4:
        return 'high rating'   
    if row.rating > 2:
        return 'average rating'
    return 'low rating'    
```

```python
ratings_movies['class'] = ratings_movies.apply(class_movies, axis=1)
ratings_movies
```

```python
movies = pd.read_csv('F:/Обучение/Профессия Аналитик данных/Python/DataSet/ml-latest/movies.csv')
movies.head()
```

```python
joined_movies_ratings = movies.merge(ratings_movies, on='movieId', how='left')
joined_movies_ratings[['title', 'rating', 'class']].sort_values('rating', ascending=False).reset_index(drop = True).head()
```

### Задание 2

Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

Результат классификации запишите в отдельный столбец region.

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
# Список городов и федеральных округов нашёл на вот этом сайте
page_url = 'https://superresearch.ru/federal-okrug'
df = pd.read_html(page_url)
city_region = df[3][[1,3]] # Нужная таблица четвертая по счету
city_region.head()
```

```python
city_region.columns = city_region.values[0] # Заменяю названия столбцов на названия, которые содержатся в первой строке
city_region = city_region[1:] # Убираю первую строку
city_region = city_region.reset_index(drop = True) # Перегружаю индекс, чтобы он как и положено начинался с 0
city_region.columns = ['Cities', 'Regions'] # Заменяю названия столбцов на английские
city_region.head()
```

```python
# Преобразую для удобства DataFrame в словарь
city_region_dict = city_region.set_index('Cities').T.to_dict('list')
city_region_dict
```

```python
# Функция для определения наименования региона по наличию указания города в тексте запроса
def contain_city(row):
    for city, region in city_region_dict.items():
        if city.lower() in row.keyword.lower():
            return region[0]
    return 'undefined'
```

```python
# Добавление наименование региона в исходный DataFrame
keywords['Region'] = keywords.apply(contain_city, axis=1)
keywords
```

### Задание 3 (бонусное)
Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.
При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:
1.	В переменную years запишите список из всех годов с 1950 по 2010.
2.	Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:  
          1.для каждой строки пройдите по всем годам списка years  
          2.если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска  
          3.если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год  
        
3.	Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’
4.	Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
years = [i for i in range(1950, 2010 + 1)]
```

```python
movies = pd.read_csv('F:/Обучение/Профессия Аналитик данных/Python/DataSet/ml-latest/movies.csv')
movies.head()
```

```python
def production_year(row):
    for year in years:
        if str(year) in row.title:
            return year
    return 1900
```

```python
# Создаю копию файла movies, чтобы не изменять оригинальный файл
movies_copy = movies.copy()
movies_copy['year'] = movies_copy.apply(production_year, axis=1)
movies_copy
```

```python
ratings = pd.read_csv('F:/Обучение/Профессия Аналитик данных/Python/DataSet/ml-latest/ratings.csv')
ratings.head()
```

```python
# Высчитываю рейтинг по каждому фильму
ratings_movies = ratings.groupby('movieId').agg({'rating':'mean'})
ratings_movies.head()
```

```python
# Создаю новый DataFrame из movies_copy и ratings_movies
joined_year_ratings = movies_copy.merge(ratings_movies, on='movieId', how='left')
# Итоговый DataFrame:
year_ratings = joined_year_ratings.groupby(['year']).agg({'rating':'mean'}).sort_values(by = 'rating', ascending = False)
year_ratings
```
